```python
# @title Utilisation de l'API II

from stackapi import StackAPI
import pandas as pd

SITE = StackAPI('stackoverflow')

# Paramètres de recherche
params = {
    'min': 50,  # score min
    #'answers': 1,  # Au moins une réponse
    'tagged': 'python',
    'pagesize': 50 ,# Nombre de résultats
    #'max_pages' : 1 #Répété 1 fois
}

#SITE.max_pages = 1
#SITE.answers = 1
questions = SITE.fetch('questions', **params)

ids = []
titles = []
tags_list = []
view_counts = []
score_counts = []
#favorite_counts = []
answer_counts = []

for question in questions['items']:
    ids.append(question['question_id'])
    titles.append(question['title'])
    tags_list.append(', '.join(question['tags']))
    view_counts.append(question['view_count'])
    score_counts.append(question['score'])
    #favorite_counts.append(question['favorite_count'])
    answer_counts.append(question['answer_count'])

data = {
    'Id' : ids,
    'Titre': titles,
    'Tags': tags_list,
    'Nombre de vues': view_counts,
    'Score' : score_counts,
    #'Nombre de marquages favoris': favorite_counts,
    'Nombre de réponses': answer_counts
}

df = pd.DataFrame(data)
df
```
